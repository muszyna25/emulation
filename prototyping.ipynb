```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_id = str(1025)

filename = exp_id + '_rmse.npz'

npz = np.load('data/' + filename)
print(npz)
```

```python
data = npz['arr_0']
#Yt = npz['arr_1']

plt.figure()
plt.semilogy(list(range(0, int(len(data)))), data, c='b')
plt.xlabel('#Iters')
plt.ylabel('RMSE')
plt.show()
```

```python
filename = exp_id + '_means_stds.npz'

npz = np.load('data/' + filename)
print(npz)

means = npz['arr_0']
stds = npz['arr_1']

plt.figure()
plt.semilogy(list(range(0, int(len(means)))), means, c='b')
plt.semilogy(list(range(0, int(len(stds)))), stds, c='g')
plt.xlabel('#Iters')
plt.ylabel('Means and Stds')
plt.show()
```

```python
filename = exp_id + '_prediction_physmodel_vs_ml.npz'

npz = np.load('data/' + filename)
print(npz)

Xt = npz['arr_0']
Yp = npz['arr_1']
Ym = npz['arr_2']

param_names = ['mu', 'sigma', 'nc', 'kappa']

for j in range(0, Xt.shape[1]):
    print(j)
    xt = Xt[:,j]

    plt.figure()

    plt.scatter(xt, Yp, c='b', label='Emulator')
    plt.scatter(xt, Ym, c='g', label='Model')

    #plt.scatter(list(range(0, Yp.shape[0])), Yp, c='b')
    #plt.scatter(list(range(0, Yp.shape[0])), Ym, c='g')
    plt.xlabel(param_names[j])
    plt.ylabel('Supersaturation')
    plt.legend()
    plt.show()
```
